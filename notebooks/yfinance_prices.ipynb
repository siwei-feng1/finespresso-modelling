```
import yfinance as yf
from datetime import datetime, timedelta
# Get today's date
#today = datetime.now()
symbol = 'TLSA'

# Calculate dates
# You can specify dates in 'YYYY-MM-DD' format
start_date = '2024-05-20'  # Example date
end_date = '2024-06-21'    # Example date

# Convert strings to datetime if needed
start_dt = datetime.strptime(start_date, '%Y-%m-%d')
end_dt = datetime.strptime(end_date, '%Y-%m-%d')

data = yf.download(symbol, interval = '1d', start = start_date, end = end_date)
data.tail(10)
```

```
import yfinance as yf
from datetime import datetime, timedelta

# Calculate dates
end_date = datetime.now()
start_date = end_date - timedelta(days=10)

# Get ticker data
ticker = yf.Ticker("TLSA")  # Example using Apple stock
df = ticker.history(
    start=start_date,
    end=end_date,
    interval='5m'
)

df.head()
```
